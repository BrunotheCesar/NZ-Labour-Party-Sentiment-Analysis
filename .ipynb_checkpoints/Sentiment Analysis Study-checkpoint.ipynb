```python
import tweepy
from textblob import TextBlob
from wordcloud import WordCloud
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
```

```python
# twitter API credentials
consumerkey = '*****'
consumersecret = '*****'
accesstoken = '*****'
accesstokensecret = '*****'
```

```python
# authentication object
authenticate = tweepy.OAuthHandler(consumerkey, consumersecret)
# set access token and secret
authenticate.set_access_token(accesstoken, accesstokensecret)
# create api object / pass auth info
api = tweepy.API(authenticate, wait_on_rate_limit = True)
```

```python
posts = api.user_timeline(screen_name ='nzlabour', count = 100, lang = 'eng', tweet_mode='extended')
#print 5 most recent tweets
print('Most recent tweets: \n')
i=1
for tweet in posts[0:5]:
    print(str(i) + ') '+ tweet.full_text + '\n')
    i = i + 1
```

```python
# create df
df = pd.DataFrame([tweet.full_text for tweet in posts], columns=['Tweets'])
print(df)
```

```python
# cleaning the text

def cleantxt(text):
    text = re.sub(r'@[A-Za-z0-9]+', '', text) #removes @
    text = re.sub(r'#', '', text) #removes #
    text = re.sub(r':', '', text) #removes :
    text = re.sub(r'RT[\s]+', '', text) #removes RT
    text = re.sub(r'^https?:\/\/.*[\r\n]*', '', text, flags=re.MULTILINE) # remove url
    text = re.sub(r'co\S+', '', text) # remove co
    text = re.sub(r'//t.', '', text)
    
    return text

df['Tweets'] = df['Tweets'].apply(cleantxt)
print(df)
    
```

```python
# subjectivity function
def getsubjectivity(text):
    return TextBlob(text).sentiment.subjectivity

# polarity function
def getpolarity(text):
    return TextBlob(text).sentiment.polarity

#adding columns to df
df['Subjectivity'] = df['Tweets'].apply(getsubjectivity)
df['Polarity'] = df['Tweets'].apply(getpolarity)

print(df)
```

```python
# word cloud
words = ' '.join([twts for twts in df['Tweets']])
wordcloud = WordCloud(width = 500, height = 300, random_state = 21, max_font_size = 119).generate(words)
plt.imshow(wordcloud, interpolation = 'bicubic')
plt.axis('off')
plt.show()
```

```python
# positive, neutral and negative function
def getanalysis(score):
    if score < 0:
        return 'Negative'
    elif score == 0:
        return 'Neutral'
    else:
        return 'Positive'
    
df['Analysis'] = df['Polarity'].apply(getanalysis)
df.head()
```

```python
# positive tweets
j = 1
sorteddf = df.sort_values(by=['Polarity'])
for i in range (0, sorteddf.shape[0]):
    if (sorteddf['Analysis'][i] == 'Positive'):
        print(str(j) + ') ' + sorteddf['Tweets'][i])
        print()
        j = j + 1
```

```python
# negative tweets
j = 1
sorteddf = df.sort_values(by=['Polarity'])
for i in range (0, sorteddf.shape[0]):
    if (sorteddf['Analysis'][i] == 'Negative'):
        print(str(j) + ') ' + sorteddf['Tweets'][i])
        print()
        j = j + 1
```

```python
# polarity and subjectivity plot
plt.figure(figsize=(8,6))
for i in range(0, df.shape[0]):
    plt.scatter(df['Polarity'][i], df['Subjectivity'][i], color='Red')

plt.title('Sentiment Analysis - NZ Labour Party Tweets')
plt.xlabel('Polarity')
plt.ylabel('Subjectivity')
plt.grid()
plt.show()
```

```python
# percentage of positive tweets

positivet = df[df.Analysis == 'Positive']
positivet = positivet['Tweets']
round((positivet.shape[0] / df.shape[0] * 100), 1)
```

```python
# percentage of negative tweets
negativet = df[df.Analysis == 'Negative']
negativet = negativet['Tweets']
round((negativet.shape[0] / df.shape[0] * 100), 1)
```

```python
# bar plot
df['Analysis'].value_counts()

plt.title('Sentiment Analysis')
plt.xlabel('Sentiment')
plt.ylabel('Count')
df['Analysis'].value_counts().plot(kind='barh')
plt.show()
```
